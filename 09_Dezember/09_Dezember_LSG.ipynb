### __1. Der HadCRUT-Datensatz__

Der globale Temperatur-Datensatz [HadCRUT](https://crudata.uea.ac.uk/cru/data/temperature/) beschreibt die **Temperaturentwicklung der Erde seit Messbeginn** im 19. Jahrhundert.

Die Datei `HadCRUT5_T_anomalies.csv` in diesem Ordner enthält Informationen über die **jährliche Temperatur seit dem Jahr 1850.** 

Wenn du darauf doppelklickst, siehst du, dass die Daten in **vier Spalten** angeordnet sind:

| JAHR  | GLOBAL | NORD | SUED |
|---|---|---|---|

Die Datei enthält also für **jedes Jahr seit 1850** Werte sowohl für die **ganze Erde** als auch für den **Nord- und Südhalbkugel.** 

In der untenstehenden Code-Zelle wird sie in ein NumPy-Array namens `daten` geladen.

```python
daten = np.loadtxt('HadCRUT5_T_anomalies.csv', skiprows=1, delimiter=',')
```


Mach dich mit den Daten vertraut: Schau dir die Variable `daten` an: überprüfe z.B. die Anzahl Reihen und Spalten in der Tabelle mit der `shape`-Eigenschaft (`daten.shape`) oder lass gleich alle Werte mit `print()` ausgeben.

```python
#lookup data
print(daten)
```

Achtung
<font size = 3> **Die Temperaturen sind als sogenannte _Anomalien_ angegeben: als Differenz zur mittleren Temperatur während der Jahre 1961-1990**. </font>

**Du musst die absoluten Temperaturen daraus berechnen!**


Die mittleren Temperaturen sind:
- 13.974°C für die ganze Erde
- 14.586°C für die nördliche Hemisphäre
- 13.361°C für die südliche Hemisphäre

und werden untenstehend als Variablen definiert.

```python
T_mean_global = 13.974
T_mean_nh     = 14.586
T_mean_sh     = 13.361
```

Du sollst **drei Abbildungen mit Zeitreihen**
erstellen (Jahr auf der $x$-Achse):
- **Abbildung 1** zeigt die Zeitreihen der Temperatur-**Anomalien** für alle drei Regionen auf einem einzigen Plot.
- **Abbildung 2** zeigt eine Zeitreihe der **absoluten, globalen** Temperatur.
- **Abbildung 3** zeigt eine Zeitreihe der **Differenz** zwischen den **absoluten** Temperaturen auf der Nord- und Südhalbkugel.


**MÖGLICHE VORGEHENSWEISE**

Als Erstes könntest du **jede Spalte von `daten` als einzelne Variable speichern**, z.B. die erste Spalte (mit Index `0`!) in eine Variable namens `jahr`.


In einem zweiten Schritt kannst du dann **alle Variablen definieren**, die du visualisieren möchten: d.h. die Anomalien in absolute Temperaturen umrechnen, und die Differenz zwischen Nord- und Südhalbkugel berechnen.

```python
#Variablen definieren
jahr = daten[:, 0] 
temp_global_anomaly = daten[:, 1]
temp_nh_anomaly = daten[:, 2]
temp_sh_anomaly = daten[:, 3]

# Absolute Temperaturen berechnen
temp_global_abs = temp_global_anomaly+T_mean_global
temp_nh_abs = temp_nh_anomaly+T_mean_nh
temp_sh_abs = temp_sh_anomaly+T_mean_sh
```

Haben du das alles in der obigen code-Zelle gemacht, musst du nur noch die drei Abbildungen erstellen!

<font size = 3> **Die folgende Abbildung zeigt sogenannte Temperatur-Anomalien (Abweichungen zum Mittel) seit 1850. Die drei Zeitreihen bilden eine Aufschlüsselung nach Region (Global, Nord- und Südhalbkugel) ab, wobei die Nordhalbkugel den höchsten Anstieg verzeichnet.** </font>

```python
#Plot initialisieren
fig, ax = plt.subplots()

#Zeitreihen für die drei Regionen erstellen
ax.plot(jahr, temp_global_anomaly , color='y', linestyle='-', label="Global")
ax.plot(jahr, temp_sh_anomaly , color='g', linestyle='-',  label="Südhalbkugel")
ax.plot(jahr, temp_nh_anomaly , color='r', linestyle='-',  label="Nordhalbkugel")

#Legende platzieren
ax.legend(loc="upper left")

#Achsen- und Titelbeschriftung
ax.set_xlabel('Zeit (Jahr)')
ax.set_ylabel('Temp (Grad Celsius)')
ax.set_title('Temperatur-Anomalien seit 1850')
fig.show()
```

<font size = 3> **Die folgende Abbildung zeigt die globale Temperaturentwicklung seit 1850. Ein Trend der globalen Erwärmung ist sichtbar.** </font>

```python
#Plot initialisieren
fig, ax = plt.subplots()

#Zeitreihen für die drei Regionen erstellen
ax.plot(jahr, temp_global_abs , color='r', linestyle='-')

#Achsen- und Titelbeschriftung
ax.set_xlabel('Zeit (Jahr)')
ax.set_ylabel('Temp. (Grad Celsius)')
ax.set_title('Globale Temperaturentwicklung seit 1850')
fig.show()
```


<font size = 3> **Die folgende Abbildung zeigt die Differenz der Temperatur zwischen Nord- und Südhalbkugel seit 1850. Es ist ersichtlich, dass sich die Differenz in den letzten Jahrzehnten erheblich erhöht hat.**</font>

```python
fig, ax = plt.subplots()

temp_diff_nh_sh = abs(temp_nh_abs - temp_sh_abs)

ax.plot(jahr, temp_diff_nh_sh , color='r', linestyle='-')
fig.show()

#Achsen- und Titelbeschriftung
ax.set_xlabel('Zeit (Jahr)')
ax.set_ylabel('Temp. (Grad Celsius)')
ax.set_title('Differenz zwischen den absoluten Tempereaturen der Nord- und Südhalbkugel seit 1850')
fig.show()
```
