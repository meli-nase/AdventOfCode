Die Datei `CH_seiscat_2014.csv` in diesem Ordner enthält Informationen über Beben **im Jahr 2014**. 

Die Daten sind in vier Spalten angeordnet:
| LON [°E]  | LAT [°N] | TIEFE [KM] | MAGNITUDE |
|---|---|---|---|

**Die Datei enthält also für jedes Erdbeben im Jahr 2014 vier Kenngrössen:**

- die [__*Magnitude*__](http://seismo.ethz.ch/de/knowledge/things-to-know/faq/#cc_00124.xml) des Bebens, ein Mass für die freigesetzte Energie
