### 1.2. Datenimport
Die Daten sind in zwei Kolonnen angeordnet : die erste Kolonne enthält die Kategorie, die zweite Kolonne die Anzahl Fans in der Familie Baumann.
### 1.3. Daten in Python laden

Jetzt wollen wir die Daten in Python laden. Das wird in der untenstehenden Code-Zelle gemacht. 

Mit der bestehenden Zeile werden die Daten in eine Variable namens `umfrage_daten` geladen. __<font color='#208de2'> Füge eine neue Zeile in die untenstehende code-Zelle ein, und tippe den Variablennamen erneut ein. Damit siehst du, welche Daten geladen wurden.

```python
umfrage_daten = np.loadtxt("pizza.csv", delimiter=',', skiprows=1, dtype='str')

print(umfrage_daten)
```


### 1.4. Daten visualisieren

Wir wollen jetzt die Daten als Bar-Chart visualisieren.

Die untenstehende __code__-Zelle teilt die Kolonnen von `umfrage_daten` in zwei Variabeln auf: 
- die Toppings kommen in eine Variable namens `toppings`
- die  Anzahl Fans pro Kategorie in die Variable `anzahl`.

```python
# Extrahiere Toppings (Spalte 0) und Scores (Spalte 1, konvertiert zu Integer)
toppings = umfrage_daten[:, 0]
anzahl = umfrage_daten[:, 1].astype(int)

# Kombiniere sie in eine Liste von Tupeln (Topping, Score)
data_tuples = list(zip(toppings, anzahl))

# Erstelle das Balkendiagramm
plt.figure(figsize=(10, 6))
bars = plt.bar(toppings, anzahl, color='skyblue')

# Setze Beschriftungen und Titel
plt.xlabel('Pizza Topping')
plt.ylabel('Begeisterte Familienmitglieder')
plt.title('Pizza Topping Präferenz')
```

Bonus: Nun noch ein bisschen schöner, bitte! Sortiert :)

```python
# Extrahiere Toppings (Spalte 0) und Scores (Spalte 1, konvertiert zu Integer)
toppings = umfrage_daten[:, 0]
anzahl = umfrage_daten[:, 1].astype(int)

# Kombiniere sie in eine Liste von Tupeln (Topping, Score)
data_tuples = list(zip(toppings, anzahl))

# Sortiere die Daten nach 'Score' (dem zweiten Element im Tupel) in absteigender Reihenfolge
data_sorted = sorted(data_tuples, key=lambda item: item[1], reverse=True)

# Trenne die sortierten Daten in separate Listen für das Plotten
toppings_sorted = [item[0] for item in data_sorted]
scores_sorted = [item[1] for item in data_sorted]

# Erstelle das Balkendiagramm
plt.figure(figsize=(10, 6))
bars = plt.bar(toppings_sorted, scores_sorted, color='skyblue')

# Setze Beschriftungen und Titel
plt.xlabel('Pizza Topping')
plt.ylabel('Begeisterte Familienmitglieder')
plt.title('Pizza Topping Präferenz')

# Rotiere die X-Achsen-Beschriftungen für bessere Lesbarkeit
plt.xticks(rotation=45, ha='right')

# Füge Gitterlinien für besseres Ablesen der Scores hinzu
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Passe das Layout an, um abgeschnittene Beschriftungen zu verhindern
plt.tight_layout()
```
