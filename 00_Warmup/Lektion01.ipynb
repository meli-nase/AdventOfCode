### 1.4. Daten visualisieren

Wir wollen jetzt die Anzahl Studierende pro Kategorie als __Kuchendiagramm__ visualisieren. 

Diese Daten befinden sich in der __zweiten Kolonne__ von `umfrage_daten`. 

:::{attention} Wichtig!
__Die Nummerierung fängt in Python immer bei 0 an__ - das heisst, die erste Kolonne ist Kolonne `0`, die zweite Kolonne `1`. 
:::

Die untenstehende __code__-Zelle teilt die Kolonnen von `umfrage_daten` in zwei Variabeln auf: 
- die Kategorienamen kommen in eine Variable namens `kategorien`
- die  Anzahl Studierende pro Kategorie in die Variable `anzahl`. 

Die Anzahl Studierende wird dann visualisiert als Kuchendiagramm (_pie chart_) mit dem Befehl `plt.pie(anzahl)`. __<font color='#208de2'> Führen Sie diese Zelle mit Shift+Enter aus. </font>__

```python
kategorien = umfrage_daten[:, 0] # die erste Kolonne von umfrage_daten wird in die Variable kategorien gespeichert
anzahl = umfrage_daten[:, 1] # die zweite Kolonne geht in die Variable anzahl

plt.bar(height=anzahl)             # die Variable anzahl wird als Kuchendiagramm dargestellt
plt.show()


farben = 'red', 'orange', 'yellow', 'green'
plt.pie(anzahl, labels=kategorien, radius=1.5, startangle= 90, colors=farben)             # die Variable anzahl wird als Kuchendiagramm dargestellt
plt.show()



'''res={}
for i in range(len(kategorien)):
  values = kategorien[i]
  keys=anzahl[i]
  dictionary = dict(zip(keys, values))
  res.update(dictionary)

print(res)

names = [key for key,value in res.values() if value!=0]
values = [value for value in res.values() if value!=0]
plt.pie(values, labels=names, radius=1.5, startangle= 90, colors=farben)
plt.show()
'''
```
